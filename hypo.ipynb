```python
df = pd.read_csv("/content/Finance_data.csv")
df.head()
```

```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind

# Step 1: Load the dataset
# Assuming your dataset is in CSV format and located at 'Finance_data.csv'
data = df


# Step 2: Group the data by gender and extract Fixed Deposits
males_fd = data[data['gender'] == 'Male']['Fixed_Deposits']
females_fd = data[data['gender'] == 'Female']['Fixed_Deposits']

# Step 3: Perform an independent t-test
t_stat, p_value = ttest_ind(males_fd, females_fd, alternative='greater')

# Step 4: Set the significance level
alpha = 0.05

# Step 5: Interpret the results
print(f"T-Statistic: {t_stat}")
print(f"P-Value: {p_value}")

if p_value < alpha:
    print("Reject the null hypothesis (H0): Males invest more in Fixed Deposits than females.")
else:
    print("Fail to reject the null hypothesis (H0): Males do not invest more in Fixed Deposits than females.")
```
